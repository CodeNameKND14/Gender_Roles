```python
import numpy as np
import pandas as pd
from password import password
import sqlalchemy as sa
from sqlalchemy import create_engine
```

### Import data from Kaggle
Source: https://www.kaggle.com/ronaldtroncoso20/sf-salaries-extended/

```python
csv_file = "Resources/SF_salary_data_gender.csv"
#sf_salary_data = pd.read_csv(csv_file,na_values = ["Not Provided", "nan"],dtype={"Status":str})
sf_salary_data = pd.read_csv(csv_file)
sf_salary_data
```

### Fix Problems with mixed Data Types within Columns

```python
df1 = sf_salary_data.replace(to_replace =["Not Provided", "nan", "not provided"], value ="NaN")
```

```python
df2 = df1.fillna({"gender":"u"})
```

```python
objects_df = df2[["Id","Employee Name","JobTitle","Notes","Agency","Status","gender"]]
```

```python
floats_df = df2.drop(["Employee Name","JobTitle","Notes","Agency","Status","gender"],axis=1).apply(lambda s: s.astype("float")).fillna(0)
new_floats_df = floats_df["Id"].astype("int")
floats_df["Id"] = new_floats_df
```

```python
master_df = pd.merge(objects_df, floats_df, on="Id")
```

### Remove Problem-cell in "Employee Name" Column with lazer beam iloc function ;)

```python
master_df.iloc[19230, 1] = "NaN"
```

### Remove two Columns that aren't useful

```python
master_df1 = master_df.drop('Notes', axis=1)
```

```python
master_df2 = master_df1.rename(columns={"Unnamed: 0": "Drop"})
```

```python
master_db = master_df2.drop("Drop", axis=1)
master_db.head()
```

### Convert "Year" Column to Integer

```python
year_fixed = master_db["Year"].astype(int)
master_db["Year"] = year_fixed
```

### Transform DF Column Names for Loading Into PostgreSQL

```python
master_cols = ["Id", "Employee Name", "JobTitle", "Agency", "Status", "gender", "BasePay", "OvertimePay", "OtherPay", "Benefits", "TotalPay", "TotalPayBenefits", "Year"]
master_transformed = master_db[master_cols].copy()

# Rename the column headers
master_transformed = master_transformed.rename(columns={"Id": "id",
                                                        "Employee Name": "employee_name",
                                                        "JobTitle": "job_title",
                                                        "Agency": "agency",
                                                        "Status": "status",
                                                        "gender": "gender",
                                                        "BasePay": "base_pay",
                                                        "OvertimePay": "overtime_pay",
                                                        "OtherPay": "other_pay",
                                                        "Benefits": "benefits",
                                                        "TotalPay": "total_pay",
                                                        "TotalPayBenefits": "total_pay_benefits",
                                                        "Year": "year"})
```

### Divide Master/Transformed DF into 4 Separate DF's by Year for Analysis in SQL Database

```python
df_2011 = master_transformed[master_transformed["year"] == 2011]
df_2012 = master_transformed[master_transformed["year"] == 2012]
df_2013 = master_transformed[master_transformed["year"] == 2013]
df_2014 = master_transformed[master_transformed["year"] == 2014]
```

### Export Finalized DF to CSV as a Data Back-Up 

```python
master_transformed.to_csv('Resources/master_db.csv')
```
