# Clean and Inspect Data

```python
# exctract everything in the ResponsiblePerson column and create two seperate columns 
# into a seperate column 
```

```python
# exctract everything in the column and make into a seperate column 
uk_gender_df["ResponsiblePerson"].value_counts()
```

```python
uk_gender_df["ResponsiblePerson"] = uk_gender_df.ResponsiblePerson.astype(str)
```

# Make Two Seperate Columns 

```python
# Convert this column to a list 
name_titles = uk_gender_df["ResponsiblePerson"].tolist()
```

```python
# Split the the name and title and place them in seperate lists
title = []
name= []
for x in name_titles:
    try:
        n,t = x.split("(")
        t.replace(")", "")
        title.append(t)
        name.append(n)
    except:
        title.append("NaN")
        name.append(n)
        
```

```python
# Turn those lists into column names 
uk_gender_df["Responsible_Person"] = name
uk_gender_df["Title"] = title
```

```python
uk_gender_df = uk_gender_df.rename(columns={"ResponsiblePerson": "name_title",
                                           "EmployerName": "employer_name",
                                            "Address": "address",
                                            "CompanyNumber": "company_number",
                                            "SicCodes": "sic_codes",
                                            "MaleBonusPercent": "male_bonus_percent",
                                            "FemaleBonusPercent": "female_bonus_percent",
                                            "MaleUpperMiddleQuartile": "male_upper_middle_quartile",
                                            "FemaleUpperMiddleQuartile": "female_upper_middle_quartile",
                                            "MaleTopQuartile": "male_top_quartile",
                                            "FemaleTopQuartile": "female_top_quartile",
                                            "Responsible_Person": "responsible_person",
                                            "Title": "title"
                                           })
```

```python
uk_gender_df.head()
```

```python
# Select the coumns I want
uk_cols = ["employer_name", "address", 
           "company_number", "sic_codes", 
           "DiffMeanHourlyPercent", "DiffMedianHourlyPercent", 
           "DiffMeanBonusPercent", "DiffMedianBonusPercent",
           "male_bonus_percent", "female_bonus_percent",
           "male_upper_middle_quartile", "female_upper_middle_quartile",
           "male_top_quartile", "female_top_quartile", 
           "CompanyLinkToGPGInfo", "responsible_person",
           "title", "name_title"
          ]
```

```python
# creat a new DataFrame with those selected columns 
uk_gender_compare_df = uk_gender_df[uk_cols].copy()
```
